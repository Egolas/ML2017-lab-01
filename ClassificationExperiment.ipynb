```python
import numpy as np
from sklearn.datasets import load_svmlight_file
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt


# define model function
def model_function(x):
    return np.dot(x, omega[:14])[0] + omega[14]


# define model function with threshold
def model_function_t(x, threshold):
    if model_function(x) > threshold:
        return 1
    else:
        return -1


# define hinge loss of i
def hinge_loss_i(given_x, given_y):
    return max(0, 1 - given_y * model_function(given_x))


# define loss function (hinge loss)
def loss_function(given_x, given_y):
    loss = 0
    for cur_x, cur_y in zip(given_x, given_y):
        cur_x = cur_x.toarray()
        loss = loss + hinge_loss_i(cur_x, cur_y)

    loss = loss / given_y.size
    return loss


# define accuracy rate funtion
def accuracy_function(given_x, given_y, threshold):
    objective_num = 0
    for cur_x, cur_y in zip(given_x, given_y):
        objective_num = objective_num + max(0, cur_y * model_function_t(cur_x.toarray(), threshold))

    return objective_num / given_y.size


# define partial derivative of hinge loss of i with respect to w
def partial_derivative_of_w(given_x, given_y):
    if (1 - given_y * model_function(given_x)) >= 0:
        return - given_y * given_x
    else:
        return 0


# define partial derivative of hinge loss of i with respect to b
def partial_derivative_of_b(given_x, given_y):
    if (1 - given_y * model_function(given_x)) >= 0:
        return - given_y
    else:
        return 0


# load data set
X, y = load_svmlight_file("australian_scale")

# split data set
X_train, X_validation, y_train, y_validation = train_test_split(X, y, test_size=0.33, random_state=42)

# parameters initialize - zeros
omega = np.zeros(15)

# define learning rate
eta = 0.1

# define iteration number
iteration_num = 200

# gradient decent
train_loss = []
validation_loss = []
acc_rate = []
for i in range(iteration_num):
    # print current iteration number
    print("\rProducing (" + str(i + 1) + "/" + str(iteration_num) + ")", end="")

    # define C
    C = 0.9

    # calculate gradient
    gradient = np.zeros(15)
    gradient_of_w = np.zeros(14)
    gradient_of_b = 0
    for cur_x, cur_y in zip(X_train, y_train):
        gradient_of_w = gradient_of_w + C * partial_derivative_of_w(cur_x.toarray(), cur_y)
        gradient_of_b = gradient_of_b + C * partial_derivative_of_b(cur_x.toarray(), cur_y)

    gradient_of_w = omega[:14] + gradient_of_w

    gradient = np.append(gradient_of_w, gradient_of_b)
    # gradient divided by number of samples
    for j in range(len(gradient)):
        gradient[j] = gradient[j] / y_train.size

    # update parameters
    omega = omega - eta * gradient

    # calculate loss
    train_loss.append(loss_function(X_train, y_train))
    validation_loss.append(loss_function(X_validation, y_validation))

    # calculate acc rate
    acc_rate.append(accuracy_function(X_validation, y_validation, 0))

# plot loss changes with iteration
# yellow line for validation data loss
plt.plot(range(iteration_num), train_loss, label="Train")
plt.plot(range(iteration_num), validation_loss, label="Validation")
plt.plot(range(iteration_num), acc_rate, label="Accuracy")
plt.legend()

plt.xlabel("Iteration number")
plt.ylabel("Loss/Accuracy")
plt.title("Linear Classification and gradient decent")
plt.show()


```
